```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('administration_data.csv')
```

```python
df = df.groupby(['gender', 'age']).agg({'comprehension': 'mean', 'production': 'mean'})
```

```python
plt.figure(figsize=(15,7))
plt.grid(axis='x', alpha=0.7)

ys = list(range(8, 37))

plt.xticks(ticks=ys[::6], fontsize=13)
plt.xlabel(', .', fontsize=15)
plt.yticks(fontsize=13)
plt.ylabel(' ', fontsize=15)

plt.plot(ys, df.loc['Female']['comprehension'].to_numpy(), label=', ')
plt.plot(ys, df.loc['Female']['production'].to_numpy(), label=', ')
plt.plot(ys, df.loc['Male']['comprehension'].to_numpy(), label=', ')
plt.plot(ys, df.loc['Male']['production'].to_numpy(), label=', ')

plt.title('       ', fontsize=18)
plt.legend(fontsize=12)
plt.show()
```
